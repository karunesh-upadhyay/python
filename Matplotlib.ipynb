
# (2) BarÂ Chart
A bar chart is a visual way of showing information using rectangular bars. Each bar's length corresponds to the value it represents, making it easy to compare different categories.

## [a] Simple Bar Plot

```python
from matplotlib import pyplot as plt
import numpy as np
plt.style.use("fivethirtyeight")

ages_x = range(25,36)
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]

# bar chart
plt.bar(ages_x,dev_y,color='#444444',label = "ALL Devs") # Median salary of all dev

py_dev_y = [45372, 48876, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
plt.bar(ages_x,py_dev_y,color='#008fd5',label = "Python") # Median salary of python dev

js_dev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.bar(ages_x,js_dev_y,color='#e5ae38',label = "Java") # Median salary of JS dev

plt.legend(loc = 'upper left')

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary-All devs (USD)")

plt.tight_layout()
plt.show()
```

**Note:** In the above chart, python & java numbers are visible but do not have any visibility for all devs(hidden behind). Hence, it would be better to have side by side bars. We cannot use plt.bar for multiple plots as it will just lay over 1 plt on another which might hide some info. We will use numpy(np) for offsetting x to do the same.
* step1: create a list ranging from 0 to length of the list
* step2: assign a width of columns (default = 0.8)
* step3: create bar plots where change the x_plot values so that all 3 can be arranged side-by-side. Add width as an argument in the plt.bar statement and adjust x_indices by adding and subtracting width.
* step4: use xticks to arrange exact values on x-axis

## [b] Side-by-Side Bar Chart/ Adjacent BarÂ Chart

```python
from matplotlib import pyplot as plt
import numpy as np
plt.style.use("fivethirtyeight")

ages_x = range(25,36)
dev_y = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]

x_indices = np.arange(len(ages_x))
width = 0.2
plt.bar(x_indices-width,dev_y,width=width,color='#444444',label = "ALL Devs") # Median salary of all dev

py_dev_y = [45372, 48876, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
plt.bar(x_indices,py_dev_y,width=width,color='#008fd5',label = "Python") # Median salary of python dev

js_dev_y = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]
plt.bar(x_indices+width,js_dev_y,width=width,color='#e5ae38',label = "Java") # Median salary of JS dev

plt.xticks( ticks = x_indices, labels=ages_x)
plt.legend(fontsize="10")

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median Salary-All devs (USD)")

plt.tight_layout()
plt.show()
```

# (3) PieÂ Plots
A pie plot is a visually appealing chart that showcases proportions within a whole by dividing a circle into slices. It's effective for illustrating the distribution of categories in a dataset, making it easier to grasp relative contributions or percentages. The size of each slice corresponds to the proportion it represents, providing an intuitive representation of categorical relationships.

## [a] Simple Pie Chart

```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")

slices = [40,60]
plt.pie(slices)
plt.title('Pie Chart')
plt.tight_layout()
plt.show()
```

## [b] DetailedÂ PieChart

```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")

slices = [40,35,25]
labels = ['forty','thirty-five','twenty-five'] # similarly can add colors = colors list
colors = ['#008fd5','#fc4f30','#e5ae37']
explode = [0,0,0.1] # 0.1 of radius exploded out

plt.pie(slices , labels = labels,
        colors = colors,
        shadow = True,
        explode = explode,
        startangle = 90,     # first division starts with 90-degree.
        autopct = '%1.1f%%',  # percentages added on chart
        wedgeprops = {'edgecolor':'black'}) # add black edges in the plot

plt.title('Pie Chart')
plt.tight_layout()
plt.show()
```

# (5) Filling Area on LineÂ Plots
We might need to highlight the area under a plot or between 2 plots or even compare a plot with respect to a line. We will cover all such variations in this section

## [a] Filling Area between Plot andÂ X-axis
* Plotting dev and py salaries with respect to age and also filling the area between the line plot of dev salary and horizontal axis.
* Note alpha is used to determine the opacity of the filling.
* Major con of this method is that the filling area might cover up other plots

```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

# X Values
ages = range(25,36)

# Y Values
dev_salaries = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
py_salaries = [45372, 48876, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
js_salaries = [37810, 43515, 46823, 49293, 53437, 56373, 62375, 66674, 68745, 68746, 74583]

# ploting dev and py salaries
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, label='Python')

# filling area between X-axis and dev salary
plt.fill_between(ages,dev_salaries, alpha=0.25) # fills till bottom and might hide other plots too. alpha changes intensity of color of fillins

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

# (6) Histograms
We can create histogram plot for values in an array. We Will cover 2 options i.e. one where we can input the total number of bins and  another where we can decide the bin boundary



## [a] Histogramâ-âTotal Number ofÂ bins
Plotting histogram for ages with total of 5 bins

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")

ages = [5,5,6,7,8,10,14,15,20,22,23,24]

# This generated frequency for values in age with total of 5 bins
# edge color used to differentiate between bins else  with same blue color
# wont be able to understand different bins
plt.hist(ages,bins=5,edgecolor='black')
plt.legend()

plt.title('Histogram bins of Age')
plt.xlabel('Ages')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

#(9) Using Subplots

Matplotlib's subplots function allows you to create multiple plots within a single figure. This is particularly useful when you want to compare different datasets or visualize related information.

**figure**: The figure is the top-level container that holds and encompasses all elements of a plot. It is the outermost container and represents the entire canvas or window. It represents the entire window or page in which the plot appears.

**axis**: An axis is a specific subplot within a figure where the data will be plotted.

**Key Differences between figure and axisÂ :**
The figure is the entire window or page, while the axis is a specific plotting area within that window.
A figure can contain one or more axes, allowing for multiple subplots in a single window.
The figure is often used for high-level customization of the entire plot (e.g., setting the title of the entire plot), while the axis is used for customizing individual subplots.

Following are the replacements required using axis

* plt.plot() => ax.plot()
* plt.legend() => ax.legend()
* plt.title() => ax.set_title()
* plt.xlabel() => ax.set_xlabel()
